```python
import numpy as np
import sys
from numpy import linalg
import matplotlib.pyplot as plt
```

```python
#gauss_jacobi - com criterio de linhas
ordem = int(input('determine a ordem  da matriz: ')) # ordem da matriz
A = np.zeros((ordem,ordem))
N = 100
for i in range(ordem):
    for j in range(ordem):
        A[i,j]=float(input('A'+str(i)+str(j)+'='))
print(A)
if np.linalg.det(A) ==0:
    sys.exit('DET =0, Matriz Singular ou possui Equações Linearmente Dependentes')
    
print('det A: ', np.linalg.det(A))

#criterio de linha
alfa=np.zeros(ordem)
for i in range(ordem):
    alfa[i]=sum(A[i])
    alfa[i]=abs((alfa[i]-A[i,i])/abs(A[i,i]))
print('alfa: ',alfa)
if max(alfa)>1:
    sys.exit('Nao atende o critério das linhas')


B = np.zeros(ordem)
for i in range(ordem):
    B[i]=float(input('B'+str(i)+'='))
print(B)

# Vetor Inicial
X0=np.zeros(ordem)
for i in range(ordem):
    X0[i]=B[i]/A[i,i]
print('X0: ',X0)
tol = float(input('determine a tol do erro: '))
it = 0  
#iteracoes jacobi
X = np.zeros(ordem)
while it<100:
    it +=1
    for i in range(ordem):
        X[i] = B[i]
        for j in range(ordem):
            if j != i:
                X[i] -= A[i,j]*X0[j]
        X[i] /= A[i,i]
    print('X'+str(it)+'= ',X)
    erro = np.linalg.norm(X-X0,np.inf)
    errorel = np.linalg.norm(X-X0,np.inf)/abs(max(X))
    print('erro: ',erro)
    print('errorel: ',errorel)
    if erro<tol or errorel<tol:
        sys.exit('Criterio de Parada Atingido')
    X0= np.copy(X) 
```

```python
#gauss_seidel- com criterio de sassenfeld
ordem = int(input('determine a ordem  da matriz: ')) # ordem da matriz
A = np.zeros((ordem,ordem))
N = 100
for i in range(ordem):
    for j in range(ordem):
        A[i,j]=float(input('A'+str(i)+str(j)+'='))
print(A)
if np.linalg.det(A) ==0:
    sys.exit('DET =0, Matriz Singular ou possui Equações Linearmente Dependentes')
    
print('det A: ', np.linalg.det(A))

#criterio de sassenfeld
beta=np.zeros(ordem)

for i in range(ordem):
    beta[i]= 0
    beta[i] = sum(A[i])-A[i][i]
    j=0
    while j < i:
        beta[i] -= A[i][j]
        beta[i] += A[i][j]*beta[j]
        j += 1
    beta[i] /= A[i][i]
print('beta: ',beta)
if max(beta)>1:
    sys.exit('Nao atende o critério de sassenfeld')


B = np.zeros(ordem)
for i in range(ordem):
    B[i]=float(input('B'+str(i)+'='))
print(B)

# Vetor Inicial
X0=np.zeros(ordem)
for i in range(ordem):
    X0[i]=B[i]/A[i,i]
print('X0: ',X0)
tol = float(input('determine a tol do erro: '))
it = 0
#iteracoes seidel
X = np.zeros(ordem)
X= np.copy(X0)
while it<100:
    it +=1
    for i in range(ordem):
        X[i] = B[i]
        for j in range(ordem):
            if j != i:
                X[i] -= A[i,j]*X[j]
        X[i] /= A[i,i]
    print('X'+str(it)+'= ',X)
    erro = np.linalg.norm(X-X0,np.inf)
    errorel = np.linalg.norm(X-X0,np.inf)/abs(max(X))
    print('erro: ',erro)
    print('errorel: ',errorel)
    if erro<tol or errorel<tol:
        sys.exit('Criterio de Parada Atingido')
    X0= np.copy(X) 
```
