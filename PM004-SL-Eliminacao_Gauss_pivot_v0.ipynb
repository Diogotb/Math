```python
import numpy as np
import sys
from numpy import linalg
import matplotlib.pyplot as plt
```

```python
#Eliminação de Gauss
ordem = int(input('determine a ordem  da matriz: ')) # ordem da matriz
A = np.zeros((ordem,ordem))
N = 100
for i in range(ordem):
    for j in range(ordem):
        A[i,j]=float(input('A'+str(i)+str(j)+'='))
print(A)
if np.linalg.det(A) ==0:
    sys.exit('DET =0, Matriz Singular ou possui Equações Linearmente Dependentes')
    
print('det A: ', np.linalg.det(A))
B = np.zeros(ordem)
for i in range(ordem):
    B[i]=float(input('B'+str(i)+'='))
print(B)
# Triangulaçao da matriz
mult = 0
for i in range(ordem-1):
    j=i+1
    while j<ordem:
        mult = A[j,i]/A[i,i]
        print('Pivot: ',A[j,i])
        print('Multiplicador: ',mult)
        for k in range(ordem):
            A[j,k] = A[j,k]/mult - A[i,k]
        B[j] = B[j]/mult - B[i]
        j +=1
    print(A)
    print(B)
    print()
# Resolução
X = np.zeros(ordem)
X[ordem-1] = B[ordem-1]/A[ordem-1,ordem-1]
i=ordem-2
while i>=0:
    sum = 0
    j = ordem-1
    while j>i:
        sum = sum + A[i,j]*X[j]
        j-=1
    X[i] = (B[i]-sum)/A[i,i]
    i-=1
    
print('resultado: ',X)
```
