```python
import numpy as np
import sys
from numpy import linalg
import matplotlib.pyplot as plt
```

```python
#Eliminação de Gauss
ordem = int(input('determine a ordem  da matriz: ')) # ordem da matriz
A = np.zeros((ordem,ordem))
N = 100
for i in range(ordem):
    for j in range(ordem):
        A[i,j]=float(input('A'+str(i)+str(j)+'='))
print(A)
if np.linalg.det(A) ==0:
    sys.exit('DET =0, Matriz Singular ou possui Equações Linearmente Dependentes')
    
print('det A: ', np.linalg.det(A))
B = np.zeros(ordem)
for i in range(ordem):
    B[i]=float(input('B'+str(i)+'='))
print(B)
# Triangulaçao da matriz
mult = 0
for i in range(ordem-1):
    j=i+1
    while j<ordem:
        mult = A[j,i]/A[i,i]
        print('Pivot: ',A[j,i])
        print('Multiplicador: ',mult)
        for k in range(ordem):
            A[j,k] = A[j,k]/mult - A[i,k]
        B[j] = B[j]/mult - B[i]
        j +=1
    print(A)
    print(B)
    print()
# Resolução
X = np.zeros(ordem)
X[ordem-1] = B[ordem-1]/A[ordem-1,ordem-1]
i=ordem-2
while i>=0:
    sum = 0
    j = ordem-1
    while j>i:
        sum = sum + A[i,j]*X[j]
        j-=1
    X[i] = (B[i]-sum)/A[i,i]
    i-=1
    
print('resultado: ',X)
```

```python
#Fatoração LU
ordem = int(input('determine a ordem  da matriz: ')) # ordem da matriz
A = np.zeros((ordem,ordem))
N = 100
for i in range(ordem):
    for j in range(ordem):
        A[i,j]=float(input('A'+str(i)+str(j)+'='))
print(A)
if np.linalg.det(A) ==0:
    sys.exit('DET =0, Matriz Singular ou possui Equações Linearmente Dependentes')    
print('det A: ', np.linalg.det(A))
B = np.zeros(ordem)
for i in range(ordem):
    B[i]=float(input('B'+str(i)+'='))
print(B)
# Triangulaçao da matriz LU
mult = 0
i=0
L = np.eye(ordem)
U = np.copy(A)
while i<ordem-1:
    j=i+1
    while j<ordem:
        mult = U[j,i]/U[i,i]
        print('Multiplicador: ',mult)
        L[j,i] = mult
        for k in range(ordem):
            U[j,k] = U[j,k] - U[i,k]*mult  
        U[j,i]=0
        j +=1
    print(L)
    print(U)
    print()
    i+=1

# Resolução
Y = np.zeros(ordem)
Y = np.linalg.solve(L,B)
print('Y: ',Y)

X = np.zeros(ordem)
X = np.linalg.solve(U,Y)
print('X: ',X)
```

```python
# gauss_jacobi - com criterio de linhas
ordem = int(input('determine a ordem  da matriz: ')) # ordem da matriz
A = np.zeros((ordem,ordem))
N = 100
for i in range(ordem):
    for j in range(ordem):
        A[i,j]=float(input('A'+str(i)+str(j)+'='))
print(A)
if np.linalg.det(A) ==0:
    sys.exit('DET =0, Matriz Singular ou possui Equações Linearmente Dependentes')
    
print('det A: ', np.linalg.det(A))

#criterio de linha
alfa=np.zeros(ordem)
for i in range(ordem):
    alfa[i]=sum(A[i])
    alfa[i]=abs((alfa[i]-A[i,i])/abs(A[i,i]))
print('alfa: ',alfa)
if max(alfa)>1:
    sys.exit('Nao atende o critério das linhas')


B = np.zeros(ordem)
for i in range(ordem):
    B[i]=float(input('B'+str(i)+'='))
print(B)

# Vetor Inicial
X0=np.zeros(ordem)
for i in range(ordem):
    X0[i]=B[i]/A[i,i]
print('X0: ',X0)
tol = float(input('determine a tol do erro: '))
it = 0  
#iteracoes jacobi
X = np.zeros(ordem)
while it<100:
    it +=1
    for i in range(ordem):
        X[i] = B[i]
        for j in range(ordem):
            if j != i:
                X[i] -= A[i,j]*X0[j]
        X[i] /= A[i,i]
    print('X'+str(it)+'= ',X)
    erro = np.linalg.norm(X-X0,np.inf)
    errorel = np.linalg.norm(X-X0,np.inf)/abs(max(X))
    print('erro: ',erro)
    print('errorel: ',errorel)
    if erro<tol or errorel<tol:
        sys.exit('Criterio de Parada Atingido')
    X0= np.copy(X) 
```
